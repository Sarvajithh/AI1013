**Average Search Delays**

```python
# Initialize the avg_search_delay list
avg_search_delay = []


for col in search_times_df.columns:
    # Convert column data to NumPy array, replacing invalid values with NaN
    oddball_data = np.array(pd.to_numeric(search_times_df[col], errors='coerce'))

    # Compute mean- this way it even ignores if there are NaN values
    oddball_mean = np.nanmean(oddball_data) - 328  # Subtract baseline search time

    # Append the values
    avg_search_delay.append(oddball_mean)

# Convert to NumPy array-as asked in the question
avg_search_delay = np.array(avg_search_delay)


avg_search_delay
```

**KL - Divergence**

```python
# Initialising the list
def cal_entropic_distances(data):
    relative_entropic_distance = []

    for i in range(0, len(data.columns), 2):
        # Labeling columns
        col_L = data.columns[i]
        col_R = data.columns[i+1]

        # Extracting required data
        lambda_i = np.array(pd.to_numeric(data[col_L][3:177], errors='coerce')) 
        lambda_j = np.array(pd.to_numeric(data[col_R][3:177], errors='coerce'))
    
        # Computing D_ij while skipping zero entries
        D_ij_values = []
        D_ji_values = []
        
        for li, lj in zip(lambda_i, lambda_j):
            if li == 0 or lj == 0:
                continue  
            
            D_ij_values.append(li * (np.log(li / lj)) - li + lj)
            D_ji_values.append(lj * (np.log(lj / li)) - lj + li)

        # Compute the mean of non-skipped values
        if D_ij_values:  # Ensure the list is not empty to avoid NaN issues
            D_ij = np.mean(D_ij_values)
            D_ji = np.mean(D_ji_values)
            relative_entropic_distance.append(D_ij)
            relative_entropic_distance.append(D_ji)

    relative_entropic_distance = np.array(relative_entropic_distance)
    return relative_entropic_distance
```

```python

relative_entropic_distance = cal_entropic_distances(firing_rates_df)
relative_entropic_distance
```

$l_1$ norm 

```python
def cal_l1_distances(data):
    # Initialize list for L1 distances
    l_1_distances = []

    # Iterate over consecutive column pairs
    for i in range(0, len(data.columns), 2):  # Step by 2 to match KL divergence pairs
        col_L = data.columns[i]  # First column in pair
        col_R = data.columns[i+1]  # Second column in pair

        # Convert to numeric and normalize (since you normalized KL, do the same here)
        lambda_i = np.array(pd.to_numeric(data[col_L], errors='coerce'))
        lambda_j = np.array(pd.to_numeric(data[col_R], errors='coerce'))

        # Remove NaN values to prevent errors
        valid_indices = ~np.isnan(lambda_i) & ~np.isnan(lambda_j)
        lambda_i = lambda_i[valid_indices]
        lambda_j = lambda_j[valid_indices]
        

        # Compute L1 distance if valid data exists
        if len(lambda_i) == 0 or len(lambda_j) == 0:
            L_ij = np.nan  # Store NaN if no valid data
        else:
            L_ij = np.mean(np.abs(lambda_i - lambda_j))

        # Append result
        l_1_distances.append(L_ij)
        l_1_distances.append(L_ij)

    # Convert to NumPy array
    l_1_distances = np.array(l_1_distances)

    # Display results
    return l_1_distances
```

```python
l_1_distances = cal_l1_distances(firing_rates_df)
l_1_distances
```

```python
inverse_search_delay = 1/np.array(avg_search_delay)
inverse_search_delay
```

To find the best fitting line through origin, we use linear regression on the data to find the slope without an intercept.
$$y = mx$$
$$\text{Error} = y-mx$$

total squared error  = $\Sigma (y_i-mx_i)^2$

Objective: To find $m$ which minimises the error

$$\frac{d(\Sigma (y_i - mx_i)^2)}{dm} = 0$$

Simplifying we get, 
$$ m = \frac{\Sigma x_i y_i}{\Sigma x_i^2}$$


**Finding the slopes of lines**

```python
kl_fitting_slope = np.dot(relative_entropic_distance, inverse_search_delay)/np.dot(relative_entropic_distance, relative_entropic_distance)
kl_fitting_points = kl_fitting_slope * np.array(relative_entropic_distance)

l1_fitting_slope = np.dot(l_1_distances, inverse_search_delay)/np.dot(l_1_distances, l_1_distances)
l1_fitting_points = l1_fitting_slope * np.array(l_1_distances)
```

**Plotting**

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(8, 6))
plt.axis((0,2,0,0.0045))
# Scatter plot for data point
plt.scatter(relative_entropic_distance, inverse_search_delay, color='green', label='Data Points')
plt.plot(relative_entropic_distance, kl_fitting_points, color='blue', linestyle='-', label=f'Best Fit: y = {kl_fitting_slope:.5f}x')
# Labels and title
plt.xlabel("Relative Entropy Distances (KL Divergence)")
plt.ylabel("Inverse Search Delay")
plt.title("KL Divergence vs Inverse Search Delay")
x = np.linspace(0,2,1000)
y = kl_fitting_slope*x
plt.plot(x,y)
plt.grid(True)
plt.legend()
plt.show()
```

**3rd- Part**

```python
import random as rd
rd.seed(1)

search_times_cols = search_times_df.columns
random_cols = rd.sample(list(search_times_cols), 9)

mean_search_delays = []
std_search_delays = []
for i in random_cols:
    search_times = np.array(pd.to_numeric(search_times_df[i][2:75], errors='coerce')) 
    search_times = search_times-328
    mean_search_delays.append(np.nanmean(search_times))
    std_search_delays.append(np.nanstd(search_times))
    
std_search_delays = np.array(std_search_delays)
mean_search_delays = np.array(mean_search_delays)

# Plottng 
fitting_slope = np.dot(std_search_delays, mean_search_delays)/np.dot(mean_search_delays, mean_search_delays)
x = np.linspace(0,1200,1200)
y = fitting_slope*x
plt.figure(figsize=(8,5))
plt.plot(x,y,'r-', label=f'Best fit: y ={fitting_slope:.5}x')
plt.grid(True)
plt.scatter(mean_search_delays, std_search_delays)
# Label
plt.xlabel('Mean of search delays')
plt.ylabel("Standard deviation of search delays")
plt.title("mean_search_delays vs std_search_delays")
plt.legend()
plt.show()

shape_param = (mean_search_delays/std_search_delays)**2
shape_estimate = np.mean(shape_param)
print(f"Estimated shape parameter is {shape_estimate}")
```

```python

import scipy.stats as stats


# (b) Select the other half of the columns
remaining_cols = list(set(search_times_df.columns) - set(random_cols))
rd.seed(1)
sampled_cols_1 = rd.sample(remaining_cols, len(remaining_cols)//2)
sample_data_mean = []
sample_data_std = []
for i in sampled_cols_1:
    sampled_search_times = np.array(pd.to_numeric(search_times_df[i][2:75], errors='coerce'))
    sampled_search_times -= 328
    sample_data_mean.append(np.mean(sampled_search_times))
    sample_data_std.append(np.mean(sampled_search_times))

rate_parameter = shape_estimate/np.mean(sample_data_mean)
print(f"Estimated rate parameter (b): {rate_parameter}")

# (c) Collect remaining half of data from the remaining columns
sampled_cols_2 = list(set(random_cols) - set(sampled_cols_1))
search_delays = []

for col in sampled_cols_2:
    remaining_data = np.array(pd.to_numeric(search_times_df[col][2:75], errors='coerce'))
    remaining_data -= 328  # Subtract 328 ms
    search_delays.extend(remaining_data)  # Append data to list

search_delays = np.array(search_delays)  # Convert to NumPy array

# Sort the search delays
search_delays_sorted = np.sort(search_delays)

# Compute empirical CDF
empirical_cdf = np.arange(1, len(search_delays_sorted) + 1) / len(search_delays_sorted)

gamma_cdf = stats.gamma.cdf(search_delays_sorted, a=shape_estimate, scale=1/rate_parameter)

ks_statistic = np.max(np.abs(empirical_cdf - gamma_cdf))


# Plotting 
print(f"Kolmogorov–Smirnov test statistic: {ks_statistic}")
plt.figure(figsize=(8, 5))
plt.plot(search_delays_sorted, empirical_cdf, label="Empirical CDF",  linestyle='-')
plt.plot(search_delays_sorted, gamma_cdf, label="Theoretical Gamma CDF", linestyle='--')
plt.xlabel("Search Delays (ms)")
plt.ylabel("CDF")
plt.title("Empirical vs Theoretical CDF (Gamma Distribution)")
plt.legend()
plt.grid()
plt.show()


```
